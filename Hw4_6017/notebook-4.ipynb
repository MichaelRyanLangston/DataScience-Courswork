## Createing data frames from csv files...

```python
import pandas as pd
import seaborn as c
import statsmodels.formula.api as sm

bucketKNNVaried_D = pd.read_csv('spatialDatastructuresStarterCode/BucketKNNVaried_D.csv', names=['k','N','D','Time'], skiprows=1)
dumbKNNVaried_D = pd.read_csv('spatialDatastructuresStarterCode/dumbKNNVaried_D.csv', names=['k','N','D','Time'], skiprows=1)
KDTreeKNNVaried_D = pd.read_csv('spatialDatastructuresStarterCode/KDTreeKNNVaried_D.csv', names=['k','N','D','Time'], skiprows=1)

KDTreeKNNVaried_k = pd.read_csv('spatialDatastructuresStarterCode/KDTreeKNNVaried_k.csv', names=['k','N','D','Time'], skiprows=1)
QuadTreeKNNVaried_k = pd.read_csv('spatialDatastructuresStarterCode/QuadTreeKNNVaried_k.csv', names=['k','N','D','Time'], skiprows=1)

KDTreeKNNVaried_N = pd.read_csv('spatialDatastructuresStarterCode/KDTreeKNNVaried_N.csv', names=['k','N','D','Time'], skiprows=1)
QuadTreeKNNVaried_N = pd.read_csv('spatialDatastructuresStarterCode/QuadTreeKNNVaried_N.csv', names=['k','N','D','Time'], skiprows=1)

KDTreeKNNConstD_2 = pd.read_csv('spatialDatastructuresStarterCode/KDTreeKNNConstD_2.csv', names=['k','N','D','Time'], skiprows=1)
QuadTreeKNNConstD_2 = pd.read_csv('spatialDatastructuresStarterCode/QuadTreeKNNConstD_2.csv', names=['k','N','D','Time'], skiprows=1)
```

## Analysis of Varied Dimension


```python
display(bucketKNNVaried_D)
c.set_theme()
c.scatterplot(x=bucketKNNVaried_D["D"], y=bucketKNNVaried_D["Time"]).set(title='KNNVaried_D')
c.regplot(x=bucketKNNVaried_D["D"], y=bucketKNNVaried_D["Time"], data=bucketKNNVaried_D)
buckReg = sm.ols(formula='Time ~ D', data=bucketKNNVaried_D).fit()
display(buckReg.summary())

display(dumbKNNVaried_D)
c.scatterplot(x=dumbKNNVaried_D["D"], y=dumbKNNVaried_D["Time"])
c.regplot(x=dumbKNNVaried_D["D"], y=dumbKNNVaried_D["Time"], data=dumbKNNVaried_D)
dumbReg = sm.ols(formula='Time ~ D', data=dumbKNNVaried_D).fit()
display(dumbReg.summary())

display(KDTreeKNNVaried_D)
c.scatterplot(x=KDTreeKNNVaried_D["D"], y=KDTreeKNNVaried_D["Time"])
c.regplot(x=KDTreeKNNVaried_D["D"], y=KDTreeKNNVaried_D["Time"], data=KDTreeKNNVaried_D)
KDTreeReg = sm.ols(formula='Time ~ D', data=KDTreeKNNVaried_D).fit()
display(KDTreeReg.summary())

```

### From the above analysis it appears the best method for increasing dimention appears to be the DumbKNN closely followed by the KDTreeKNN. The bucketKNN suffers the most from the increase in dimension.


## Analysis of Varied k Neighbors


```python
display(KDTreeKNNVaried_k)
c.set_theme()
c.scatterplot(x=KDTreeKNNVaried_k["k"], y=KDTreeKNNVaried_k["Time"]).set(title='KNNVaried_k')
c.regplot(x=KDTreeKNNVaried_k["k"], y=KDTreeKNNVaried_k["Time"], data=KDTreeKNNVaried_k)
KDTreeReg = sm.ols(formula='Time ~ k', data=KDTreeKNNVaried_k).fit()
display(KDTreeReg.summary())

display(QuadTreeKNNVaried_k)
c.scatterplot(x=QuadTreeKNNVaried_k["k"], y=QuadTreeKNNVaried_k["Time"])
c.regplot(x=QuadTreeKNNVaried_k["k"], y=QuadTreeKNNVaried_k["Time"], data=QuadTreeKNNVaried_k)
QuadReg = sm.ols(formula='Time ~ k', data=QuadTreeKNNVaried_k).fit()
display(QuadReg.summary())
```

## From the above analysis it apperas that the QuadTree has a better KNN method than the KD tree as far as speed is concerend, perhaps due to less tree traversal when considering 2 dimentions.

## Analysis of Varied N (Number of training Points)

```python
display(KDTreeKNNVaried_N)
c.set_theme()
c.scatterplot(x=KDTreeKNNVaried_N["N"], y=KDTreeKNNVaried_N["Time"]).set(title='KNNVaried_N')
c.regplot(x=KDTreeKNNVaried_N["N"], y=KDTreeKNNVaried_N["Time"], data=KDTreeKNNVaried_N)
KDTreeReg = sm.ols(formula='Time ~ N', data=KDTreeKNNVaried_N).fit()
display(KDTreeReg.summary())

display(QuadTreeKNNVaried_N)
c.scatterplot(x=QuadTreeKNNVaried_N["N"], y=QuadTreeKNNVaried_N["Time"])
c.regplot(x=QuadTreeKNNVaried_N["N"], y=QuadTreeKNNVaried_N["Time"], data=QuadTreeKNNVaried_N)
QuadReg = sm.ols(formula='Time ~ N', data=QuadTreeKNNVaried_N).fit()
display(QuadReg.summary())
```

## From the data shown above for a large number of points, KNN queries appear to be faster when appling the the QuadTree's method.

## Analysis of Constand Dimension 2

```python

print('KDTree Analysis')
display(KDTreeKNNConstD_2)
KDTreePlot = c.pairplot(data=KDTreeKNNConstD_2).set(title='KDTreeKNNConstD_2')
KDTreeReg = sm.ols(formula='Time ~ k + N', data=KDTreeKNNConstD_2).fit()
display(KDTreeReg.summary())

print('QuadTree Analysis')
display(QuadTreeKNNConstD_2)
quadTreePairPlot = c.pairplot(data=QuadTreeKNNConstD_2).set(title='QuadTreeKNNConstD_2')
QuadReg = sm.ols(formula='Time ~ k + N', data=QuadTreeKNNConstD_2).fit()
display(QuadReg.summary())
```

## So from the above we can see that the increase in time is dependent on both k and N as expected. Since the Quad Tree seems to out perform the KD tree when the dimension is at a fixed value of 2, it would be a better choice to performe and store the data in that type of tree. This result was somewhat expected since the advantage of the KD tree is its independence from dimensionality as the dimesion increases in complexity. We can see this advantage when the KDTree was compared with the bucketKNN method. The only item better in speed was the dumbKNN and this was perhaps due to the implementation of the KDTree.
